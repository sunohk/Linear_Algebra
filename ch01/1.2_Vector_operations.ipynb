# Adding vectors

```python
v = np.array([4,5,6])
w = np.array([10,20,30])
u = np.array([0,3,6,9])

vPlusW = v + w
# uPlusW = u + w # 오류(차원 불일치)

print(v)
print(w)
print(vPlusW)
```

# Vector-scalar multiplication

```python
# a scalar
s = -1/2

# a vector
b = np.array([3,4])  #b가 리스트이면 오류 발생(리스트는 정수로만 정수만큼 반복하기 떄문)

print(b*s)
```

```python
# Effects of different scalars

# a list of scalars:
scalars = [ 1, 2, 1/3, 0, -2/3 ]

baseVector = np.array([ .75,1 ])

# create a figure
fig,axs = plt.subplots(1,len(scalars),figsize=(12,3))
i = 0 # axis counter

for s in scalars:

  # compute the scaled vector
  v = s*baseVector

  axs[i].arrow(0,0,baseVector[0],baseVector[1],head_width=.3,width=.1,color='k',length_includes_head=True)
  axs[i].arrow(.1,0,v[0],v[1],head_width=.3,width=.1,color=[.75,.75,.75],length_includes_head=True)
  axs[i].grid(linestyle='--')
  axs[i].axis('square')
  axs[i].axis([-2.5,2.5,-2.5,2.5])
  axs[i].set(xticks=np.arange(-2,3), yticks=np.arange(-2,3))
  axs[i].set_title(f'σ= {s:.2f}')
  i+=1 # update axis counter

plt.tight_layout()
plt.savefig('Figure_01_03.png',dpi=300)
plt.show()

     
```

# Vector transpose

```python
# Row vector
r = np.array([[1,2,3]])

print(r), print('   ')
print(r.T), print('   ')
print(r.T.T)
```

```python
l = np.array([1,2,3]) #orientationless(방향이 없는)

print(l), print('   ')
print(l.T), print('   ')
print(l.T.T)
```

```python
v = np.array([[1,2,3]]).T # column vector
w = np.array([[10,20]]) # row vector

print(v+w)
```

```python
v = np.array([[1,2,3]]) # row vector
w = np.array([[10,20]]).T # column vector

print(v+w)
```

# The dot product is distributive

```python
# just some random vectors
v = np.array([0,1,2])
w = np.array([3,5,8])
u = np.array([13,21,34])

# two ways to compute
res1 = np.dot(v, w+u)
res2 = np.dot(v,w) + np.dot(v,u) #분배 법칙

# show that they are equivalent(결과 동일)
res1, res2 
```

# Other vector product

- Hardamard product(아다마르 곱) : 두 벡터의 대응되는 각 원소 곱하기

```python
a = np.array([5,4,8,2])
b = np.array([1,0,.5,-1])

a * b
```

- Outer product(외적) : 두 벡터를 곱해서 행렬을 생성

```python
outer_product = np.outer(a,b)
outer_product
```
