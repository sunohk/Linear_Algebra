```python
# Effects of different scalars

# a list of scalars:
scalars = [ 1, 2, 1/3, 0, -2/3 ]

baseVector = np.array([ .75,1 ])

# create a figure
fig,axs = plt.subplots(1,len(scalars),figsize=(12,3))
i = 0 # axis counter

for s in scalars:

  # compute the scaled vector
  v = s*baseVector

  axs[i].arrow(0,0,baseVector[0],baseVector[1],head_width=.3,width=.1,color='k',length_includes_head=True)
  axs[i].arrow(.1,0,v[0],v[1],head_width=.3,width=.1,color=[.75,.75,.75],length_includes_head=True)
  axs[i].grid(linestyle='--')
  axs[i].axis('square')
  axs[i].axis([-2.5,2.5,-2.5,2.5])
  axs[i].set(xticks=np.arange(-2,3), yticks=np.arange(-2,3))
  axs[i].set_title(f'σ= {s:.2f}')
  i+=1 # update axis counter

plt.tight_layout()
plt.savefig('Figure_01_03.png',dpi=300)
plt.show()

     
```

# The dot product is distributive

```python
# just some random vectors
v = np.array([0,1,2])
w = np.array([3,5,8])
u = np.array([13,21,34])

# two ways to compute
res1 = np.dot(v, w+u)
res2 = np.dot(v,w) + np.dot(v,u) #분배 법칙

# show that they are equivalent(결과 동일)
res1, res2 
```

# Other vector product

## Exercise 3

```python
def createUnitVector(v):
    mu = np.linalg.norm(v)
    return v / mu # 단위 벡터 출력

w = np.array([0,1,0])
print(createUnitVector(w))

w = np.array([0,3,0])
print(createUnitVector(w))

w = np.array([13,-5,7])
uw = createUnitVector(w)
print(uw), print('  ')
print(np.linalg.norm(w), np.linalg.norm(uw))

print('\n\n\n')
createUnitVector(np.zeros((4,1)))
```

```python
def createUnitVector(v):
    mu = np.linalg.norm(v)
    if mu == 0: #0벡터 예외 처리
        return np.zeros_like(v) #0벡터 그대로 반환(기존 입력한 배열과 같은 모양과 데이터 타입 유지)
    return v / mu # 단위 벡터 출력

createUnitVector(np.zeros((4,1))) #np.zeros_like와 달리 shape을 직접 지정해야 함
```

## Exercise 4

```python
def createMagVector(v, mag):
    mu = np.linalg.norm(v)
    return mag*v/mu

w = np.array([1,0,0])
mw = createMagVector(w,4)
print(mw)

print(np.linalg.norm(w), np.linalg.norm(mw))
```

## Exercise 9

```python
t = np.random.randn(2)
r = np.random.randn(2)

# decomposition(분해)
t_para = r*(np.dot(t,r)/np.dot(r,r)) # 평행 성분(parallel)
t_perp = t - t_para # 수직 성분(perpendicular)

print(t)
print(t_para+t_perp)

print(np.dot(t_para, t_perp)) #서로 직교하므로 값은 0(-1.1058862159352145e-17)

plt.figure(figsize=(6,6))

plt.plot([0, t[0]], [0, t[1]], color='k', linewidth=3, label=r't')
plt.plot([0, r[0]], [0, r[1]], color=[.7,.7,.7], linewidth=3, label=r'r')

plt.plot([0, t_para[0]], [0, t_para[1]], 'b--', linewidth=3, label='t_para')
plt.plot([0, t_perp[0]], [0, t_perp[1]], 'g--', linewidth=3, label='t_perp')

plt.axis('equal')
plt.legend()
plt.savefig('Figure_01_08.png', dpi=600)
plt.show()
```
