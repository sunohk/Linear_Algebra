```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg') 
plt.rcParams.update({'font.size':14})
plt.rcParams["font.family"] = "Times New Roman"
```

# Linear weighted combinations

```python
l1 = 1
l2 = 2
l3 = -3

v1 = np.array([4,5,1])
v2 = np.array([-4,0,-4])
v3 = np.array([1,3,2])

l1*v1 + l2*v2 + l3*v3
```

# Basis vectors and points

```python
p = (3,1)
q = (-6,2)

plt.figure(figsize=(6,6))

plt.plot(p[0], p[1], 'ko', markerfacecolor='b', markersize=10, label='Point p')
plt.plot(q[0], q[1], 'ks', markerfacecolor='g', markersize=10, label='Point q')

plt.plot([0,0], [0,1], 'k', linewidth=3, label='Basis S')
plt.plot([0,1], [0,0], 'k', linewidth=3)

plt.plot([0,3], [0,1], 'k--', linewidth=3, label='Basis T')
plt.plot([0,-3], [0,1], 'k--', linewidth=3)

plt.axis('square')
plt.grid(linestyle='--', color=[.8,.8,.8])
plt.xlim([-7, 7])
plt.ylim([-7, 7])
plt.legend()
plt.savefig('Figure_02_04.png', dpi=600)
plt.show()
```
