# Creating vectors

```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg') #display figures in vecotr format
plt.rcParams.update({'font.size' : 14}) # set global font size
plt.rcParams["font.family"] = "Times New Roman"
```

```python
asList = [1,2,3]

asArray = np.array([1,2,3]) #1차원 배열(방향이 없는 배열 = 1차원 리스트)

rowVec = np.array([ [1,2,3] ])
colVec = np.array([[1],[2],[3]])
```

```python
print(f'asList :  {np.shape(asList)}')
print(f'asArray : {asArray.shape}')
print(f'rowVec : {rowVec.shape}')
print(f'colVec : {colVec.shape}')
```

# Geometry of vectors

```python
# create a vector
v = np.array([-1,2])

#plot that vector (and a dot for the tail)

plt.arrow(0,0,v[0], v[1], head_width=.5, width=.1) # x좌표, y좌표, 화살표 시작 지점으로부터 x축으로 이동한 값, 화살표 시작 지점으로부터 y축으로 이동한 값, 화살표 head의 두께, 화살표 두께
plt.plot(0,0, 'ko', markerfacecolor='k', markersize=7) # 'ko' : k=black, o : 원형 마커 , markerfacecolor : 마커 내부 색상

#add axis lines
plt.plot([-3,3], [0,0], '--', color=[.8, .8, .8], zorder=-1) #수평 점선 #x좌표 목록(-3~3까지), zorder : 그래프 요소의 순서(맨뒤, 맨앞 등)를 결정하는 값(값이 클수록 앞(위), 값이 작을수록 뒤(아래) 위치)
plt.plot([0,0], [-3,3], '--', color=[.8, .8, .8], zorder=-1) #수직 점선 


plt.axis('square')
plt.axis([-3,3,-3,3])
plt.xlabel('v0')
plt.ylabel('v1')
plt.title('Vector v in standard position')
plt.show()
```

```python
# A range of starting positions

startPos = [
            [0,0],
            [-1,-1],
            [1.5,-2]
            ]

#create a new figure

fig = plt.figure(figsize=(6,6))

for s in startPos:

    #plot that vector (and a dot for the tail)
    #note that plt.arrow automatically adds an offset to the third/fourth inputs
    plt.arrow(s[0], s[1], v[0], v[1], head_width=.5, width=.1, color='black')
    plt.plot(s[0], s[1], 'ko', markerfacecolor='r', markersize=7)

    #indicate the vector in its standard position
    if s==[0,0]:
        plt.text(v[0]+.1, v[1]+.2, '"Standard position"') #기준 위치

#add axis lines
plt.plot([-3,3], [0,0], '--', color=[.8, .8, .8], zorder=-1)
plt.plot([0,0], [-3,3], '--', color=[.8, .8, .8], zorder=-1)

plt.axis('square')
plt.axis([-3,3,-3,3])
plt.xlabel('v0')
plt.ylabel('v1')
plt.title('Vector v in various locations')
plt.savefig('Figure_01_01.png', dpi=600) #write out the fig to a file #dpi : 1인치당 300개의 점(픽셀)을 사용하여 이미지를 저장, 값이 클수록 이미지가 더 선명
plt.show()
```
