```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg') 
plt.rcParams.update({'font.size':14})
plt.rcParams["font.family"] = "Times New Roman"
```

# Linear weighted combinations

```python
l1 = 1
l2 = 2
l3 = -3

v1 = np.array([4,5,1])
v2 = np.array([-4,0,-4])
v3 = np.array([1,3,2])

l1*v1 + l2*v2 + l3*v3
```

# Basis vectors and points

```python
p = (3,1)
q = (-6,2)

plt.figure(figsize=(6,6))

plt.plot(p[0], p[1], 'ko', markerfacecolor='b', markersize=10, label='Point p')
plt.plot(q[0], q[1], 'ks', markerfacecolor='g', markersize=10, label='Point q')

plt.plot([0,0], [0,1], 'k', linewidth=3, label='Basis S')
plt.plot([0,1], [0,0], 'k', linewidth=3)

plt.plot([0,3], [0,1], 'k--', linewidth=3, label='Basis T')
plt.plot([0,-3], [0,1], 'k--', linewidth=3)

plt.axis('square')
plt.grid(linestyle='--', color=[.8,.8,.8])
plt.xlim([-7, 7])
plt.ylim([-7, 7])
plt.legend()
plt.savefig('Figure_02_04.png', dpi=600)
plt.show()
```

## Exercise 1

```python
scalars = [l1, l2, l3]
vectors = [v1, v2, v3]

#initialize the linear combination
linCombo = np.zeros(len(v1))

for s, v in zip(scalars, vectors): # (scalars, vectors) 형태로 묶어서 반복
    linCombo += s*v

linCombo
```

## Exercise 2

```python
# vector count > scalar count --> 결과 오류 발생(서로 길이가 다르기 때문)

scalars = [l1, l2, l3, 5]
vectors = [v1, v2, v3]

linCombo = np.zeros(len(v1))

for i in range(len(scalars)):
    linCombo += scalars[i]*vectors[i]
```

## Exercise 3

```python
A = np.array([1,3])

xlim = [-4,4]

scalars = np.random.uniform(low=xlim[0], high=xlim[1], size=100) #np.random.uniform : 정규 분포, 균등 분포 생성(0~1 사이 값을 균등하게 size만큼 출력)

plt.figure(figsize=(6,6))

for s in scalars:
    p = A*s

    plt.plot(p[0], p[1], 'ko')

plt.xlim(xlim)
plt.ylim(xlim)
plt.grid()
plt.text(-4.5, 4.5, 'A', fontweight='bold', fontsize=18)
plt.savefig('Figure_02_07a.png', dpi=600)
plt.show()
```

```python
import plotly.graph_objects as go

v1 = np.array([3,5,1])
v2 = np.array([0,2,2])

scalars = np.random.uniform(low=xlim[0], high=xlim[1], size=(100,2)) #2차원(배열 형태)

points = np.zeros((100,3)) #3차원
for i in range(len(scalars)):
    points[i,:] = v1*scalars[i,0] + v2*scalars[i,1]

fig = go.Figure(data=[go.Scatter3d(x=points[:,0], y=points[:,1], z=points[:,2], mode='markers', marker=dict(size=6, color='blue'))])

fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
plt.savefig('Figure_02_07b.png', dpi=600)
fig.show()
```
